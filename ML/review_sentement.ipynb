```python
import pickle 
import pandas as pd
import numpy as np
from tqdm import tqdm
import time
```

```python
large = pd.read_csv("large.csv")
large
```

```python
len(large["appid"].unique())
```

```python
temp = []
for i in  tqdm(large["appid"].unique()):
    votes_dict = {}
    votes = large[large['appid']==i]["voted_up"]
    votes_dict["votes_up"] = list(votes)
    votes_dict["appid"] = i
    temp.append(votes_dict)
```

```python
data = pd.DataFrame(temp)
```

```python
positive = []
negative = []
for i in range(len(data)):
    total_votes = len(data["votes_up"][i])
    no_of_positive = sum(data["votes_up"][i])
    no_of_negative = total_votes-no_of_positive
    positive.append((no_of_positive/total_votes)*100)
    negative.append((no_of_negative/total_votes)*100)
```

```python
data["positive"] = positive
data["negative"] = negative
```
